```python
# convert date datatype from <object> to <date> type
raw["date"] = pd.to_datetime(raw["date"])
raw
```

```python
# [Pandas] How many total pageview events did the links in the provided dataset receive in the full period, how many per day?
def q1_function():
    df_1 = raw.copy()
    # return df_1.groupby( by = ["date","event"])["event"].count()
    print("How many total pageview events did the links in the provided dataset receive in the full period, how many per day?")
    pageview_cnt = df_1.loc[df_1["event"] == "pageview"]["event"].count()
    print("pageview count: ", pageview_cnt)
    date_cnt = df_1["date"].nunique()
    print("total date count: ", date_cnt)

    print( "pageview events per day: ", int(pageview_cnt / date_cnt) )
q1_function()
```

```python
#[Pandas] What about the other recorded events?
def q2_function():
    df = raw.copy()
    df_event = pd.DataFrame(df[["event"]].groupby(by = ["event"])["event"].count()).rename( columns= {"event" : "events_cnt"}).reset_index()
    df_event["date_cnt"] = pd.DataFrame(df[["event","date"]].groupby(by = "event")["date"].nunique()).reset_index()[["date"]]
    df_event["event_per_date"] = round(df_event["events_cnt"] / df_event["date_cnt"])


    return df_event
q2_function()
```
